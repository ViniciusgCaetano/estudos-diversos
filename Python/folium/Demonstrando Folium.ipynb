```python
import pandas as pd
import folium
from random import randint
```

Criando dataframe com os estados nos quais a Nagro atua e o n√∫mero de cidades abrangidas
em cada um deles

```python
dict_dados = {'Estado': ['BA', 'ES', 'GO', 'MA', 'MG', 'MS', 'MT', 'PI'
                         ,'PR', 'RS', 'SC', 'SP', 'TO'], 
              'Numero_Cidades': [randint(20, 50) for x in range(13)]}
df = pd.DataFrame(dict_dados)
df
```

### Criando mapa do Brasil'

```python
mapa_brasil = folium.Map(location=[-15.793889, -47.882778], zoom_start=3.5, tiles='Stamen Toner')
```

```python
#Criando camada de preenchimento
folium.Choropleth(
    geo_data='https://raw.githubusercontent.com/fititnt/gis-dataset-brasil/master/uf/geojson/uf.json',
    name='choropleth',
    data=df,
    columns=['Estado','Numero_Cidades'],
    key_on='properties.UF_05',
    legend_name='N√∫mero de cidades nos quais a Nagro atua',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
).add_to(mapa_brasil)

# Visualizar
mapa_brasil
```
